```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
#reading dataset skip from 1 to 90
dataset = pd.read_csv("CPIAustralia.csv",skiprows=range(1,90))
```

```python
#top 5 row of dataset
dataset.head(5)
```

```python
#Total len of rowe
len(dataset)
```

```python
#here in this relation after 1970 data is more linear
sbn.lineplot(data=dataset,x="Year",y="Australia")
```

```python
#why after 90 becouse data is new and more linear 
#Features
features=np.array(dataset[["Year","Qtr"]])[90:]


#Labels 
Australia=np.array((dataset["Australia"]))[90:]
Sydney=np.array((dataset[" Sydney"]))[90:]
Brisbane=np.array((dataset["Brisbane"]))[90:]
Adelaide=np.array((dataset["Adelaide"]))[90:]
Perth=np.array((dataset["Perth"]))[90:]
Hobart=np.array((dataset["Hobart"]))[90:]
Canberra=np.array((dataset["Canberra"]))[90:]


#Label List 
colum = ["Australia",
" Sydney",
"Brisbane",
"Adelaide",
"Perth",
"Hobart",
"Canberra"]
```

```python
# Australia
Axtrain,Axtest,Aytrain,Aytest = train_test_split(features,Australia)

# Sydney

Sxtrain,Sxtest,Sytrain,Sytest = train_test_split(features,Sydney)

# Brisbane
Bxtrain,Bxtest,Bytrain,Bytest = train_test_split(features,Brisbane)

# Adelaide
ADxtrain,ADxtest,ADytrain,ADytest = train_test_split(features,Adelaide)


# Perth
Pxtrain,Pxtest,Pytrain,Pytest = train_test_split(features,Perth)

# Hobart
Hxtrain,Hxtest,Hytrain,Hytest = train_test_split(features,Hobart)

# Canberra
Cxtrain,Cxtest,Cytrain,Cytest = train_test_split(features,Canberra)
```

```python

for each in colum:
    slope, intercept = np.polyfit(dataset["Year"], dataset[each], 1)
    x_line = np.linspace(dataset["Year"].min(), dataset["Year"].max(), 100)
    y_line = slope * x_line + intercept
    plt.scatter(dataset["Year"], dataset[each])
    plt.plot(x_line, y_line, color='red', label='Best Fit Line')
    plt.xlim(dataset["Year"].min(), dataset["Year"].max())
    plt.ylim(0, dataset[each].max())
    plt.xlabel('Year')
    plt.ylabel(each)
    plt.legend()
    plt.show()
```

```python
#just for fun it's not work with Qtr 
for each in colum:
    slope, intercept = np.polyfit(dataset["Qtr"], dataset[each], 1)
    x_line = np.linspace(dataset["Qtr"].min(), dataset["Qtr"].max(), 100)
    y_line = slope * x_line + intercept
    plt.scatter(dataset["Qtr"], dataset[each])
    plt.plot(x_line, y_line, color='red', label='Best Fit Line')
    plt.xlim(dataset["Qtr"].min(), dataset["Qtr"].max())
    plt.ylim(0, dataset[each].max())
    plt.xlabel('Qtr')
    plt.ylabel(each)
    plt.legend()
    plt.show()
```

```python
# Australia 
Lmodel=LinearRegression()
Lmodel.fit(Axtrain,Aytrain)
pred = Lmodel.predict(Axtest)
out = Lmodel.predict([Axtest[1]])



print(f"the value in features = { Axtest[1]} and model predicted = {out}")
print( f"mean of actutal test values = { Aytest.mean() } "  )
print( f"mean of predicted = { pred.mean()}" )

#check model validation 
mse =mean_squared_error(Aytest, pred)
rmse = np.sqrt(mse)
r2 = r2_score(Aytest, pred)
print(f"mean squad error { mse } ")
print(f"R mean squad error { rmse } ")
print(f"R Score =  { r2 } ")
```
